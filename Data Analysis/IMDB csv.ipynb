```python
#Classify Movies Based on Ratings [Excellent, Good, and Average]
averagethres = 4
goodthres = 7
def classify_rating(rating):
    if rating < averagethres:
        return 'Average'
    elif rating < goodthres and rating >averagethres:
        return 'Good'
    else:
        return 'Excellent'
    
df['Rating_Classification'] = df['Rating'].apply(classify_rating)
```

```python
df['Rating_Classification'].value_counts()
```
