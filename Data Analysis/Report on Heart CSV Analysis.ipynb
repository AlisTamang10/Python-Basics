## Data Description of columns
1. age
2. sex
3. cp: chest pain type (4 values)
        Value 0: typical angina
        Value 1: atypical angina
        Value 2: non-angina pain
        Value 3: asymptomatic
4. trestbps: resting blood pressure(in mm Hg on admission to the hospital
5. chol: serum cholestoral in mg/dl
6. fbs: fasting blood sugar > 120 mg/dl(1=true; 0=false)
7. restecg: resting electrocardiographic results (values 0,1,2)
        Value 0: normal
        Value 1: having ST-T wave abnormality (T-wave inversions and/or ST elevation or depression of>0.05 mV)
        Value 2: Showing probable or definite left ventricular hypertrophy by Estes criteria
8. thalach: maximum heart rate achieved
9. exang: exercise induced angina
10. oldpeak: oldpeak = ST depression induced by exercise relative to rest
11. slope: the slope of the peak exercise ST segment
        value 1: upsloping
        value 2: flat
        value 3: downsloping
12. ca: number of major vessels (0-3) colored by flourosopy
13. thal: 
        0 = normal
        1 = fixed defect
        2 = reversable defect
14. target: 
        0 = less chance of heart attack 
        1 = more chance of heart attack

## 2. Data Visualization 

### Correlation
Correlation refers to the statistical relationship between the two variables/features. It measures the extent to which two variables are linearly related to eachother. For example, there is two variables study_time and marks of any student , if we look positive correlation between these two vairables students who spends more time in study their marks tend to increase. On the other hand in negative correlation the students who spend more time in study their marks tend to decrease. 

```python
correlation = df.corr() #corr() function gives the correaltion of all dataset
print('Lets see correaltion among all variable that the dataset contained.')
correlation 
```

```python
plt.figure(figsize = (20,5))
sns.heatmap(correlation, annot = True) # annot adds numbers on heatmap to make analysis more easier
```

A correlation heatmap is a graphical tool that displays the correlation between multiple variables as a color-coded matrix. 
The color of each cell represents the strength and direction of the correlation, with less darker colors indicating stronger correlations. 
According to visualization above,
1. Negative correlations (when one variable increases, the other variable tends to decrease) indicates darker color like black for example:- age and thalach (maximum heart rate achived) with the number -0.4 .
2. Positive correlations (when one variable increases, the other variable tends to increase) indicates less darker color for example ( age and age varaible has positive correaltion or target(chances of heart attack) and chest pain has postive correlation in number 0.43)

Difference between the above two displot:-
1. The first approach bins assigned and creates a histogram with user defined bins with the kde plot at the top.
2. The second approach (kind='kde') directly creates a KDE plot without displaying a histogram.
3. The first approach gives more control over the binning of the data, while the second approach provides probable data distribution without user define boundries or groups.
